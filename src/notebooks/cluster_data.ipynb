```python
from kmodes.kprototypes import KPrototypes
import pandas as pd
import numpy as np
import os
```

```python
link_data_train = '../../data/train_data/phase-3/prob-2/train_x.parquet'
link_data_test = '../../data/captured_data/phase-3/prob-2/'
```

```python
def combine_parquet_test(url_test, parquet=False):
    raw_test = pd.DataFrame()
    count_ = 0
    for file_path in os.listdir(url_test):
        if 'parquet' not in file_path or "123.parquet" in file_path:
            continue
        captured_data = pd.read_parquet(f'{url_test}{file_path}')
        if parquet:
            captured_data['parquet_idx'] = count_
        raw_test = pd.concat([raw_test, captured_data])
        count_ += 1
    # print(count_)
    return raw_test.reset_index(drop=True)
```

```python
data_train = pd.read_parquet(link_data_train)
data_test = combine_parquet_test(link_data_test)
feature_columns = data_train.columns.tolist()
# feature_columns.remove('label')
data_test = data_test[feature_columns]
```
