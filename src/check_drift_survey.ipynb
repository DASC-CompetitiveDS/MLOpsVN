```python
import drift_survey
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
test = drift_survey.drift_survey_processing("phase-3", "prob-1", "../")
```

```python
test
```

```python
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%".format(pct, absolute)
```

```python
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
explode = (0.0, 0.0, 0.0, 0.3, 0.2, 0.0)
colors = ( "orange", "cyan", "brown",
          "grey", "indigo", "beige")
fig, ax = plt.subplots(1, 3, figsize =(20, 4))
data_1, label_1 = list(test['feature2']['train'].values()), list(test['feature2']['train'].keys())
ax[0].pie(data_1, colors = colors, explode = explode, shadow = True, startangle = 90, wedgeprops = wp, textprops = dict(color ="magenta"))
ax[0].legend(label_1,
          title ="Values",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
data_2, label_2 = list(test['feature2']['pred_non_drift']['values'].values()), list(test['feature2']['pred_non_drift']['values'].keys())
ax[1].pie(data_2, colors = colors, explode = explode, shadow = True, startangle = 90, wedgeprops = wp, textprops = dict(color ="magenta"))
ax[1].legend(label_2,
          title ="Values",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
data_3, label_3 = list(test['feature2']['pred_drift']['values'].values()), list(test['feature2']['pred_drift']['values'].keys())
ax[2].pie(data_3, colors = colors, explode = explode, shadow = True, startangle = 90, wedgeprops = wp, textprops = dict(color ="magenta"))
ax[2].legend(label_3,
          title ="Values",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
ax[0].set_title('Feature2 - Train data')
ax[1].set_title(f"Non - Drift Predition Data - {test['feature2']['pred_non_drift']['records']} batchs")
ax[2].set_title(f"Drift Predition Data - {test['feature2']['pred_drift']['records']} batchs")
```
